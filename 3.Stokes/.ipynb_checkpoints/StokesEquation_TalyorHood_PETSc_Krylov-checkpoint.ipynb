```python
from dolfin import *
import matplotlib.pyplot as plt
import time
```

```python
# Load mesh
L = 5
N = 4*2**L
mesh = UnitSquareMesh(N, N)
plot(mesh)

# Build function space
P2 = VectorElement("Lagrange", mesh.ufl_cell(), 2)
P1 = FiniteElement("Lagrange", mesh.ufl_cell(), 1)
TH = P2 * P1
W = FunctionSpace(mesh, TH)
```

```python
# Boundaries
def right(x, on_boundary): return x[0] > (1.0 - DOLFIN_EPS)
def left(x, on_boundary): return x[0] < DOLFIN_EPS
def top_bottom(x, on_boundary):
    return x[1] > 1.0 - DOLFIN_EPS or x[1] < DOLFIN_EPS

# No-slip boundary condition for velocity
noslip = Constant((0.0, 0.0))
bc0 = DirichletBC(W.sub(0), noslip, top_bottom)
bc1 = DirichletBC(W.sub(0), noslip, left)
bc2 = DirichletBC(W.sub(0), noslip, right)

# Collect boundary conditions
bcs = [bc0, bc1, bc2]
```

```python
# Define variational problem
(u, p) = TrialFunctions(W)
(v, q) = TestFunctions(W)
f = Expression(("cos(2*pi*x[0])*sin(2*pi*x[1])", "-sin(2*pi*x[0])*cos(2*pi*x[1])"), degree=2)
a = inner(grad(u), grad(v))*dx + div(v)*p*dx + q*div(u)*dx
L = inner(f, v)*dx

U = Function(W)
```

```python
# Form for use in constructing preconditioner matrix
b = inner(grad(u), grad(v))*dx + p*q*dx
```

```python
# Assemble system
A, bb = assemble_system(a, L, bcs)
```

```python
# Assemble preconditioner system
P, btmp = assemble_system(b, L, bcs)
```

```python
# Create Krylov solver and AMG preconditioner
solver = KrylovSolver("minres", "amg")

# Associate operator (A) and preconditioner matrix (P)
solver.set_operators(A, A)
```

```python
t = time.time()
solver.solve(U.vector(), bb)
t_elapsed = time.time()-t
print(t_elapsed)
```

```python
# solver.solve(U.vector(), bb)
t = time.time()
solve(a == L, U, bcs=bcs)
t_elapsed = time.time()-t
print(t_elapsed)
```

```python
LHS = PETScMatrix()
LHS = assemble(a, tensor=LHS)

RHS = PETScVector()
RHS = assemble(L, tensor=RHS)

A_mat = LHS.mat()
b_vec = RHS.vec()

from petsc4py import PETSc
myviewerM = PETSc.Viewer().createASCII(
    "LHS.txt", comm= PETSc.COMM_WORLD)
myviewerV = PETSc.Viewer().createASCII(
    "RHS.txt", comm= PETSc.COMM_WORLD)
A_mat.view(myviewerM)
b_vec.view(myviewerV)
```

```python
# Get sub-functions
u, p = U.split()

# Save solution in VTK format
ufile_pvd = File("velocity.pvd")
ufile_pvd << u
pfile_pvd = File("pressure.pvd")
pfile_pvd << p
```
